## Question 4&5

```python
# Creating a loyalty flag for existing customers using the transform() and loc() functions.
# Step one: Create new column with max order
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(15)
```

```python
# Step two: Create loyalty flag column based on max order criteria
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head(15)
```

```python
# Count number of customers by loyalty flag
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check basis statistics of product prices for each loyalty category
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

## Question 6

```python
# Target customers based on spending
# Step One: Create column of mean price by user
df_ords_prods_merge['mean_order'] = df_ords_prods_merge.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head(20)
```

```python
# Step two: Assign spending flag based on average price

df_ords_prods_merge.loc[df_ords_prods_merge['mean_order']<10, 'spending_flag'] = 'Low spender'
df_ords_prods_merge.loc[df_ords_prods_merge['mean_order']>=10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
df_ords_prods_merge[['customer_id', 'mean_order', 'spending_flag']].head(100)
```

## Question 7

```python
# Create order frequency flag based on median days since previous order.
# Step one: Create a new column showing median days since previous order.
df_ords_prods_merge['median_days_since_previous_order'] = df_ords_prods_merge.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head()
```

```python
# Step two: Assign frequency flag based on median days since previous order
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_previous_order']>20, 'Frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_previous_order']>10) & (df_ords_prods_merge['median_days_since_previous_order']<=20), 'Frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_previous_order']<=10, 'Frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge['Frequency_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge[['customer_id', 'median_days_since_previous_order', 'Frequency_flag']].head(100)
```

```python
# Renaming mean_order column to mean_price to make it easier to understand
df_ords_prods_merge.rename(columns = {'mean_order':'mean_price'}, inplace = True)
```
